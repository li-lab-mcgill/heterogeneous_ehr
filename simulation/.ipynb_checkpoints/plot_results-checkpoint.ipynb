## This notebook contains the code to produce box-plots of correlation between phi and theta matrices of the inferred and ground-truth. Each section corresponds to a particular plot as mentioned in the header text. All the plots are saved within the 'plots' folder in the present directory

```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# the correlation files obtained by running 'find_pearson_correlation_phi.py' and 'find_pearson_correlation_theta.py'
pat_results = pkl.load(open("results/patients_corr_new.pkl","rb"))
tok_results = pkl.load(open("results/tokens_corr_new.pkl","rb"))
vocab_results = pkl.load(open("results/vocab_corr_new.pkl","rb"))
types_results = pkl.load(open("results/types_corr_new.pkl","rb"))
drop_results = pkl.load(open("results/drop_corr_new.pkl","rb"))
```

### Varying number of patients

```python
pats_x_ticks = pat_results.keys()
pats_y = []
pats_y2 = []
for k, v in pat_results.items():
    pats_y.append(v[1])
    pats_y2.append(v[2])
```

```python
plt.figure(figsize=[6,5])
plt.boxplot(pats_y,showfliers=False,whis=(10,90))
#plt.boxplot(pats_y2)
plt.xticks(range(1,4),pats_x_ticks)
plt.xlabel("Number of patients")
plt.ylabel("Pearson correlation for phi")
plt.title("Number of tokens=1500, vocabulary size=2500 \nnumber of note-types=2")
plt.savefig("plots/patients_corr_new.jpg",bbox_inches="tight")
plt.show()
```

### Comparison with baseline

```python
types_x_ticks1 = [x - 0.5 for x in types_results_theta1.keys()]
types_x_ticks2 = [x for x in types_results_theta1.keys()]

types_y1 = []
for k, v in types_results_theta1.items():
    types_y1.append(v)
    
types_y2 = []
for k, v in types_results_theta2.items():
    types_y2.append(v)
```

```python
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(6, 7))
bplot1 = ax1.boxplot(types_y1[1:],whis=(10,90),showfliers=False, positions=types_x_ticks1[1:], widths=0.4,patch_artist=True)
bplot2 = ax1.boxplot(types_y2[1:],whis=(10,90),showfliers=False, positions=types_x_ticks2[1:], widths=0.4,patch_artist=True)

plt.xticks([2, 4],types_x_ticks2[1:])
plt.xlabel("Number of note-types")
plt.ylabel("Pearson correlation for theta")
plt.title("Number of patients=4000, number of tokens=1500, \nvocabulary size=2500")
colors1 = ['lightblue', 'lightblue']
colors2 = ['pink','pink','pink']
colors3 = ['green','green']
ax1.yaxis.grid(True)
for patch, color in zip(bplot1['boxes'], colors1):
    patch.set_facecolor(color)
    
for patch, color in zip(bplot2['boxes'], colors2):
    patch.set_facecolor(color)
    
ax1.legend([bplot1["boxes"][0], bplot2['boxes'][0]], ['our method', 'baseline'], loc='lower center')
plt.savefig("plots/notetypes_corr_theta_b1.pdf",bbox_inches='tight')
plt.show()
```
